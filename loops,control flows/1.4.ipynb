This section contains all about looping,for,while and loop control statements 

```python
# looping using range 
range(5)
```

```python
for i in range(5):
    print(i)
```

```python
range(1,10)
```

```python
for i in range(1,10):
    print(i)
```

```python
range(1,10,2)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(10,1,-1):
    print(i)
```

```python
# alterating over strings 
name="Abdullah"
for i in name:
    print(i)
```

```python
#while loop 
count=0
while count<5:
    print(count)
    count=count+1
```

```python
# loop control statements 
# break ends the loop premetuarly
count=0
while count<5:
    if count==3:
        break
    print(count)
    count=count+1
```

```python
# continue skips current iteration
count=0
while count<=10:
    if count%2==0:
        continue
    else:
        print(count)
    count=count+1
```

```python
#pass does nothing 
for i in range(10):
    if i==3:
        pass
    print(i)
```

```python
for i in range(10):
    if i==3:
        continue
    print(i)
```

```python
# nested looping
for i in range(3):
    for j in range(2):
        print(f"i is {i} and j is {j}")
```

```python
# calculating sum of first n natural numbers using for loop
# suppose n is 10
sum=0
for i in range(10):
    sum=sum+i

print("total sum is ",sum)
```
